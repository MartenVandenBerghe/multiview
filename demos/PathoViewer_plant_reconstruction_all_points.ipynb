```python
# Functie om overeenkomende coördinaten te vinden
def vind_overeenkomende_coordinaten(data1, data2):
    overeenkomende_coordinaten = []
    
    # Loop door de eerste dataset (data1)
    for item1 in data1:
        letter1 = item1[0]
        coord1 = (item1[1], item1[2])
        
        # Loop door de tweede dataset (data2)
        for item2 in data2:
            letter2 = item2[0]
            coord2 = (item2[1], item2[2])
            
            # Controleer of de letters overeenkomen
            if letter1 == letter2:
                # Voeg overeenkomende letter en coördinaten toe aan de lijst
                overeenkomende_coordinaten.append([letter1, coord1[0], coord1[1], coord2[0], coord2[1]])
                break  # Stop met zoeken zodra een overeenkomst is gevonden

    return overeenkomende_coordinaten

# Zoek naar overeenkomende coördinaten
resultaten = vind_overeenkomende_coordinaten(data_im1, data_im2)

data_im1_im2 = []
for letter, x1, y1, x2, y2 in resultaten:
    data_im1_im2.append([letter, x1, y1, x2, y2])

# Druk de nieuwe array af
print("Array 'data_im1_im2' met overeenkomende coördinaten:")
print("Letter", "X_im1", "Y_im1", "X_im2", "Y_im2")
for row in data_im1_im2:
    print(row)
```

```python
def find_pinholes_projections(
    filesImages,
    fileMtx,
    fileDist,
    pixel_width,
    verbose=False
):
    # Load calibration data
    mtx = np.load(fileMtx)
    dist = np.load(fileDist)
    
    # Center of image in pixels
    center = [mtx[0, 2], mtx[1, 2]]

    # Focal distance in mm
    focal_dist = pixel_width * (mtx[0, 0] + mtx[1, 1]) / 2

    # Initialize arrays to store results
    pinholes = []
    projections = []

    # Loop through each image data
    for data in filesImages:
        letter, x_im1, y_im1, x_im2, y_im2 = data
        
        # Calculate pinhole world coordinates
        if np.mean([x_im1, x_im2]) < center[0]:
            if np.mean([y_im1, y_im2]) > center[1]:
                pinhole = [-40, -65, 0]
            else:
                pinhole = [40, -65, 0]
        else:
            if np.mean([y_im1, y_im2]) > center[1]:
                pinhole = [-40, 65, 0]
            else:
                pinhole = [40, 65, 0]
        
        pinholes.append(pinhole)

        # Calculate world coordinates of projections of corners
        projections_img = []
        for x, y in [(x_im1, y_im1), (x_im2, y_im2)]:
            projection = [
                pinhole[0] + pixel_width * (center[1] - y),
                pinhole[1] - pixel_width * (center[0] - x),
                -focal_dist
            ]
            projections_img.append(projection)
        
        projections.append(projections_img)

        if verbose:
            print(f"Processing letter {letter}")

    return np.array(pinholes), np.array(projections)
```

```python
# Define the list of image data
filesImages = data_im1_im2

# Specify file paths and pixel width
fileMtx = 'D:/2024_Marten-Thesis/Calibration/mtx.npy'
fileDist = 'D:/2024_Marten-Thesis/Calibration/dist.npy'
pixel_width = 6.12e-3  # Assuming pixel width in mm

find_pinholes_projections(filesImages,fileMtx,fileDist,pixel_width)

print("Pinhole Coordinates:")
print(pinholes)
print("\nProjections Coordinates:")
print(projections)
```
